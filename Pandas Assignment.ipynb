Q10. How do you merge two Pandas DataFrames on a specific column?

```python
df11 = pd.DataFrame({'Name':['Raju', 'Rani', 'Geeta', 'Sita', 'Sohit'],
                    'Marks':[80, 90, 75, 80, 80]})
  

df12 = pd.DataFrame({'Name':['Raju', 'Divya', 'Geeta', 'Sita'],
                    'Grade':['A', 'A', 'B', 'A'],
                    'Rank':[3, 1, 4, 2 ],
                    'Gender':['Male', 'Female', 'Female', 'Female']})
print(df11)
print()
print(df12)
# first merge

df11.merge(df12[['Name','Grade','Rank']],on='Name')
```

```python
# second merge
df11.merge(df12[['Name','Grade','Rank']],on='Name',how='left')
```

```python
# third merge
df11.merge(df12[['Name','Grade','Rank']],on='Name',how='right')
```

Q24. How do you remove duplicates from a DataFrame?

```python
# check for all rows all values
display(newdf.drop_duplicates())
```

```python
# =>subset: Subset takes a column or list of column label. Itâ€™s default value is none. After passing columns,
#     it will consider them only for duplicates. 
# =>keep: keep is to control how to consider duplicate value. It has only three distinct value and default
#       is â€˜firstâ€™. 
#       If â€˜firstâ€˜, it considers first value as unique and rest of the same values as duplicate.
#       If â€˜lastâ€˜, it considers last value as unique and rest of the same values as duplicate.
#       If False, it consider all of the same values as duplicates
# =>inplace: Boolean values, removes rows with duplicates if True.

display(newdf.drop_duplicates(subset="Type 1",keep='first', inplace=True))
newdf
```
